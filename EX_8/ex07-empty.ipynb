The "flights.csv" dataset contains information about the flights of an airport. This dataset includes information such as departure and arrival time, delays, flight company, flight number, flight origin and destination, flight duration, distance, hour and minute of flight, and exact date and time of flight. This data cannot be used in management analysis and strategies and provide useful information about the performance of flights and placement companies.

The analysis of the data in this dataset can be used as a basis for the following activities:

1. Analysis of time patterns and trends: by examining the departure and arrival time of the aircraft, changes and time changes, patterns and trends in flight behavior can be identified.
2. Analysis of American companies: By viewing information about airlines such as the number of flights, the impact and overall performance, you can compare and analyze the performance of each company.
3. Analysis of delays and service quality: By examining delays and arrival time, I can collect and analyze information about the quality of services provided by the airport and companies.
4. Analysis of flight routes: by checking the origin and destination of flights, distances and flight duration, popular routes and people's choices can be identified and analyzed.
5. Analysis of airport performance: by observing the characteristics of flights and airport performance, it is possible to identify and analyze the strengths and weaknesses of the airport and suggest improvements.

It provides various tools for data analysis and visualization and can be used as a basis for managerial decisions in the field of aviation industry.

##  Distribution chart of categorical column values

```python
categorical_columns = ['carrier' ,'origin' , 'dest']


for column in categorical_columns:
    plt.figure(figsize=(20,10))
    df[column].value_counts().plot(kind='bar')
    plt.xlabel(column)
    plt.ylabel('Counts')
    plt.title(f'Distribution of {column}')
```

```python
delay = df.groupby(['origin']).agg({'dep_delay' : 'mean'})
delay.plot(kind='bar');
```

```python
delay = df.groupby(['carrier']).agg({'dep_delay' : 'mean'})
delay.plot(kind='bar');
```

```python
delay = df.groupby(['carrier']).agg({'arr_delay' : 'mean'})
delay.plot(kind='bar');
```

<p style="background-color:#4B0082;font-family:newtimeroman;color:#FFF9ED;font-size:150%;text-align:center;border-radius:10px 10px;">  Distribution chart of numerical column values </p>

## Histogram graphs of numerical values

```python
numerical_columns = ['dep_delay' , 'arr_delay' , 'air_time' , 'distance']

for column in numerical_columns :
    plt.figure(figsize=(20,10))
    sns.histplot(data=df , x=column , bins=200)
    plt.xlabel(column)
    plt.ylabel('Counts')
    plt.title(f'Distribution of {column}')
```

## Display a box plot to check the distribution and dispersion of numerical values:

```python
for column in numerical_columns :
    plt.figure(figsize=(20,10))
    sns.boxplot(data=df , x=column , showfliers=False)
    plt.xlabel(column)
    plt.ylabel('Counts')
    plt.title(f'box of {column}')
```

<p style="background-color:#4B0082;font-family:newtimeroman;color:#FFF9ED;font-size:150%;text-align:center;border-radius:10px 10px;">Analysis of relationships between variables </p>

## Display the correlation diagram (heatmap) between numerical variables:

```python
plt.figure(figsize=(20,10))

corr = df[numerical_columns].corr()

sns.heatmap(corr , annot=True)

plt.title('correlation between numerical columns')

plt.show()
```

## A diagram of the relationship between numerical variables

```python
sns.pairplot(df, vars=numerical_columns)
plt.show()
```

## Analysis of the distribution of the number of flights by month

```python
df['month'] = pd.to_datetime(df['month'], format='%m')
monthley_flights = df['month'].value_counts().sort_index()

plt.figure(figsize=(10,6))
monthley_flights.plot(marker='o');
plt.xlabel('Month')
plt.ylabel('Number of flights')
plt.title('Distribution of the number of flights by month')
plt.grid(True)
plt.show()
```

## Analysis of the distribution of the number of flights by year

```python
df['year'] = pd.to_datetime(df['year'], format='%Y')
monthley_flights = df['year'].value_counts().sort_index()

plt.figure(figsize=(10,6))
monthley_flights.plot(marker='o');
plt.xlabel('year')
plt.ylabel('Number of flights')
plt.title('Distribution of the number of flights by year')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(20,10))
sns.violinplot(data=df, x='carrier', y='dep_delay');
plt.xlabel('carrier')
plt.ylabel('dep_delay')
plt.show()
```

```python
avg_delay_Company = df.groupby(['carrier']).agg({'dep_delay': 'mean', 'arr_delay': 'mean'})
avg_delay_Company.plot(kind = 'barh');
```
