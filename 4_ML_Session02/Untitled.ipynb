```python
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
X = np.linspace(0,5,11)
y = X ** 2
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(X,y)

ax.set_title('Plot X , y')
plt.xlabel('X')
plt.ylabel('Y')
plt.xlim(1,4)
plt.ylim(1,4)
```

```python
fig = plt.figure(figsize=(8,3))
ax1 = fig.add_axes([0,0,1,1])
ax2= fig.add_axes([1.1, 0.6, 0.4, 0.4])
ax1.plot(X,y)
ax2.plot(y,X)
```
