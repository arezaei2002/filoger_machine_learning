**Import Matplotlib Module**

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

**The Data**

```python
import numpy as np
x=np.linspace(0,5,11)
y=x**2
```

```python
x
```

```python
y
```

# 1- Creating Figure

```python
# step1: create a figure canvas
fig= plt.figure(figsize=(8,3), dpi=100)

# step2: Add axes to figure
ax = fig.add_axes([0, 0, 0.8, 0.8])

# step3: Plot on the axes
ax.plot(x,y)

ax.set_title('Data Science Bootcamp', fontsize=20)
ax.set_xlabel('Time', fontsize=20)
ax.set_ylabel('Speed')
#ax.set_xlim(1,4)
#ax.set_ylim(5,15)

#plt.show()
```

**Adding another set of axes to the figure**

```python
fig = plt.figure(figsize=(8,3))

ax1= fig.add_axes([0, 0, 1, 1])
ax2= fig.add_axes([1.1, 0.6, 0.4, 0.4])


ax1.plot(x,y)
ax2.plot(y,x)

ax1.set_title('Larger Plot')
ax1.set_xlabel('MMM')

ax2.set_title('Smaller Plot')
ax2.set_xlabel('BBB')


fig.savefig('Filoger.jpg', bbox_inches='tight')
```

# 2- Creating Subplots

**Example1:**

```python

fig, ax= plt.subplots(nrows=2, ncols=2)

ax[0, 0].plot(x, x)
ax[0, 0].set_title('x vs x')
ax[0, 0].set_ylabel('Temp')

ax[0, 1].plot(x, y)
ax[1, 0].plot(y, x)
ax[1, 1].plot(x, np.sqrt(x))


fig.suptitle('Compariosn Study', fontsize=15)

plt.tight_layout()
plt.show()
```

**Example2: (with more details)**

```python
fig, ax= plt.subplots(nrows=2, ncols=2)

ax[0, 0].plot(x, x)
ax[0, 0].set_title('x vs x')
ax[0, 0].set_ylabel('Temp')

ax[0, 1].plot(x, y)
ax[1, 0].plot(y, x)
ax[1, 1].plot(x, np.sqrt(x))


fig.suptitle('Compariosn Study', fontsize=15)

fig.subplots_adjust(wspace=0.9, hspace=0.7)



#plt.tight_layout()
plt.show()
```

# 3- Matplotlib Styling

**A- Legend**

```python
fig= plt.figure()

ax=fig.add_axes([0, 0, 1, 1])

ax.plot(x, y, label='one')
ax.plot(y, x, label='two')

ax.legend(loc=(0.8, 0.5))
```

**B- Visual Styling**

```python
fig= plt.figure()

ax= fig.add_axes([0, 0, 1, 1])

ax.plot(x, x, color='#f5b402', label='x vs x', lw=5, ls='--')
ax.plot(x, x+1, label='x vs x+1' ,marker='o', markerfacecolor='r', markersize=10,
        markeredgewidth=3, markeredgecolor='g')

ax.legend()
```

# 4-Bar Plot

```python
Name=['Ali', 'Reza', 'SARA', 'MEHDI', 'MINA']
Score=[12, 16, 19, 14, 10]

fig= plt.figure()

ax= fig.add_axes([0, 0, 1, 1])

ax.bar(Name, Score)
ax.set_xlabel('Studends Name')
ax.set_ylabel('Math Score')

plt.show()
```
