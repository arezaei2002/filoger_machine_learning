```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df= pd.read_csv('dm_office_sales.csv')
```

```python
stu= pd.read_csv('StudentsPerformance.csv')
```

```python
tips = sns.load_dataset('tips')
```

# 1-Distribution

## Histogram 

```python
sns.displot(df['salary'] , color='b')
plt.title('Histogram')
plt.xlabel('Salary');
```

```python
sns.displot(df['sales']);
```

# Rug plot

```python
sns.rugplot(df['salary'] , height=0.5);
```

```python
sns.rugplot(df['sales'] , height=0.5);
```

# Kde plot

```python
sns.kdeplot(df['salary'] , label="salary")
sns.kdeplot(df['sales'] , label="sales")

plt.legend();
```

# 2-relationship plots

## scatter plot

```python
plt.figure(figsize=(8, 8), dpi=200)
sns.scatterplot(data=df , x='salary' , y='sales' , hue='level of education' , style='level of education' , palette='Set1')
```

## joinplot

```python
sns.jointplot(data=df , x='salary' , y='sales' , kind='kde' , dropna=True);
```

## pairplot

```python
sns.pairplot(data=df);
```

```python
sns.pairplot(data=stu , hue='gender');
```

# 3-Categorical plot

## barplot

```python
sns.barplot(x='sex' , y='total_bill' , data=tips)
```

## countplot

```python
sns.countplot(x='sex' , data=tips)
```

## Boxplot

```python
sns.boxplot(data=tips , x='sex' , y='total_bill')
```

```python
sns.boxplot(data=tips , x='day' , y='total_bill' , hue='smoker')
```

## violin plot

```python
sns.violinplot(data=tips , x='sex' , y='total_bill' , hue='smoker')
```

```python
sns.violinplot(data=tips , x='day' , y='total_bill' , hue='smoker')
```

```python
sns.violinplot(data=tips , x='day' , y='total_bill' , hue='sex' , split=True)
```

## strip plot

```python
sns.stripplot(data=tips , x='sex' , y='total_bill')
```

```python
sns.stripplot(data=tips , x='day' , y='total_bill')
```

```python
sns.stripplot(data=tips , x='day' , y='total_bill' , hue='sex' , dodge=True)
```

## swarm plot

```python
sns.swarmplot(data=tips , x='day' , y='total_bill')
```

```python
sns.violinplot(data=tips, x='day', y='total_bill')
sns.swarmplot(data=tips, x='day', y='total_bill', color='c')
```
