**Import all necessary Libraries**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

**Import the DataSet**

```python
df= pd.read_csv('dm_office_sales.csv')
```

**Data Overview**

```python
df.head()
```

```python
df.info()
```

# 1- Distribution Plot

### A- distplot

```python
plt.figure(figsize=(8,3))
sns.displot(data=df, x='salary', bins=15, color='r', edgecolor='b')

plt.title('Histogram')
plt.xlabel('Salary')


plt.savefig('H')
```

### B- rugplot

```python
sns.rugplot(data=df, x='salary', height=0.2)
```

### C- kdeplot

```python
sns.kdeplot(data=df, x='salary', label='Salary' )
sns.kdeplot(data=df,x='sales', label= 'Sales')

plt.legend(loc=(1.1, 0.5))
```

# 2- Relationship Plots

### A- scatter plot

```python
plt.figure(figsize=(8, 8), dpi=200)
sns.scatterplot(data=df, x='salary', y='sales', hue='level of education',style='level of education', palette='Set1')
```

### B- jointplot

```python
sns.jointplot(data=df, x='salary', y='sales')
```

```python
sns.jointplot(data=df, x='salary', y='sales', kind='kde')
```

## 3-Categorical Plot

### A-Bar Plot

```python
tips=sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.barplot(x='sex', y='total_bill', data=tips)
```

```python
sns.barplot(x='total_bill', y='sex', data=tips)
```

### B-Count Plot

```python
sns.countplot(data=tips, x='sex')
```

### C- Box Plot

```python
sns.boxplot(data=tips, x='day', y='total_bill', hue='smoker')
```

### D- Violin Plot

```python
sns.violinplot(data=tips, x='day', y='total_bill', hue='sex', split=True)
```

### E- Strip Plot

```python
sns.stripplot(data=tips, x='day', y='total_bill', hue='sex', dodge=True)

#plt.legend(loc=(1.1, 0.5))
```

### F- Swarm Plot

```python

sns.swarmplot(data=tips, x='day', y='total_bill', color='g')
```

```python
sns.violinplot(data=tips, x='day', y='total_bill')
sns.swarmplot(data=tips, x='day', y='total_bill', color='g')
```

# 4- Matrix Plot

### A- heatmap

```python
flight= sns.load_dataset('flights')
```

```python
flight.head()
```

```python
flight.shape
```

```python
fp=flight.pivot_table(index='month', columns='year', values='passengers')
```

```python
fp
```

```python
sns.heatmap(fp, cmap='Wistia', linecolor='black', linewidths=2)
```

### B- Cluster map

```python
sns.clustermap(fp)
```
